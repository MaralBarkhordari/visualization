```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import scipy
```

```python
# Scatter Plot
# Create a scatter plot to visualize random points with `x` as sequential values and `y` as random integers
np.random.seed(613)
X = np.arange(50)
Y = np.random. randint(0,100, 50)
fig, ax = plt. subplots(figsize=(5,3))
ax.scatter(X, Y)
```

```python
# Bar Plot
# Create a bar plot to visualize random integers (`y`) as bar heights for sequential `x` values fig, ax = plt.subplots (figsize=(5,3))
ax.bar (X,Y)
```

```python
# Histogram
# Generate a histogram to show the distribution of random integers (`y`)
fig, ax = plt.subplots (figsize= (5,3))
ax.hist(Y)
```

```python
# Line Plot with Title and Labels
# Create a line plot of `y` against `x`, adding a title and axis labels for context
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(X, Y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total Growth')
ax.set_xlabel('Years since start')
```

```python
# Line Plot with Font Styling
# Customize the font of the title and axis labels for a line plot of `y` against `x` 
font2 = {'family': 'monospace', 'color': 'green', 'size': 14}

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(X, Y)

ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
```

```python
# Scatter Plot with Custom Marker and Color
# Create a scatter plot using custom markers (`*`) and the color `indigo`

fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(X, Y, marker='*', color='indigo')
```

```python
# Line Plot with Custom Marker, Line Style, and Color
# Create a line plot with custom markers (`*`), dashed line style, and line width 
ax.plot(X, Y, marker='*', color='indigo', linestyle='--', linewidth=2)
```

```python
# Line Plot with Custom Marker Edge and Face Colors
# Command: Create a line plot with custom marker styles, including orange edges and black fill for markers
ax.plot(X, Y, marker='*', color='indigo', linestyle='--', linewidth=2,
        markeredgecolor='#FA9359', markerfacecolor='#000000')
```
